```python
# Eulor method python program

# function to be solved
def f(x,y):

    # uncomment below equation to solve article 01
    return 4 * y - 1 

    # uncomment below equation to solve article 02
    # return (  2*x*y / (1 + (x**2))  ) * -1

    # uncomment below equation to solve article 03
    # return x - (y**2)

    # uncomment below equation to solve article 04
    # return (-1)*(x**2)*y

    # uncomment below equation to solve article 05
    # return 2*x*(y**2)





# Euler method 
def euler(x0 , y0 , xn , n):
    
    # calculating step size
    h = (xn - x0) /n
    
    print('\n--------Solution---------')
    print('-------------------------------------------------------------')
    print('round  \t     x0  \t     y0   \t     slope   \t  n')

    for i in range(n):
        slope = f(x0, y0)
        yn = y0 + h* slope
        print('%d  \t  %.4f \t %.4f \t  %.4f \t %.4f' %( i + 1 ,x0 ,y0 , slope , yn)  )
        print('-------------------------------------------------------------')
        y0 = yn
        x0 = x0 + h

    print('\n At x = %.4f , y = %.4f' %(xn , yn))


# inputs 

print('Enter initial conditions: ')
x0 = float(input('x0 = '))
y0 = float(input('y0 = '))
print('------------------------')
10
print('Enter calculation points:')
xn = float(input('xn = '))
print('------------------------')

print('Enter number of steps :')
step = int(input('Number of steps = '))
print('------------------------')                    
    
# Euler method call
euler(x0,y0,xn,step)

#------------------------------
# Calling  function
#------------------------------
#-------------------------------
# input for article 01
#euler(0,1,0.5,10)
#------------------------------
# input for article 02
# euler(0,1,1,10)
#------------------------------
# input for article 03
# euler(0,2,0.5,10)
#------------------------------
# input for article 04
# euler(0,1,1,5)
#------------------------------
# input for article 05
# euler(0,0.5,1,10)

```
