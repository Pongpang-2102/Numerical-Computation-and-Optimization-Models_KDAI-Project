```python
# Homework Article 01
 
#-----------------------------
#   a + b + c +  d = 0   ---- Eq_01
# -2a - b + c + 2d = 1/h ---- Eq_02
#  4a + b   c + 4d = 0   ---- Eq_03

# this case , element d will be neglected (set d = 0)

# we need to start by reformatting 3 eq above into matrix from Ax = B and finding x from matrix_x = matrix (A^-1)* matrix B
#-------------------------

# h must be a constant , this case we will be using h as below value
h = 0.1
#-------------------------------------------------------------
# 01 Start with Creating array

# creating matrix A in a size of 3 x 3

eq01_A =  [1 ,  1, 1 ]
eq02_A =  [-2, -1, 1 ]
eq03_A =  [ 4,  1, 1 ]


matrix_A = np.array([ eq01_A,  eq02_A,  eq03_A  ] )

# to view inversed matrix A
matrix_A_inverse = np.linalg.inv(matrix_A )

#-----------------------------------------------------------------
# to create matrix_B

matrix_B = np.array([0, (1/h) , 0 ])
#-----------------------------------------------------------------

# to create matrix x from matrix_x = matrix (A^-1)* matrix B
matrix_x = np.array( ( matrix_A_inverse * matrix_B) )

#---------------------------------------------------------------
# to extranct array into individual value of a,b,c
row_vector = matrix_x.reshape(1,9)
a = row_vector[0,1]
b = row_vector[0,4]
c = row_vector[0,7]


```

```python
# to view info from matrix A
print("matrix A (A) = \n")
print(matrix_A)
print("\n---------------------------------")

# to view inversed matrix A
print("matrix A inversed (A^-1) = \n")
print(matrix_A_inverse)
print("\n---------------------------------")

# to view matrix B
print("matrix B (B) = \n")
print(matrix_B)
print("\n---------------------------------")

# to view matrix x
print("matrix x can be calculated as below equation....\n")
print(" from B = Ax    \n      x  = (A^-1) * B \n")
print("matrix x (x) = \n")
print(matrix_x)
print("\n---------------------------------")


# to extract non-zero element in matrix_x and to be individual a,b,c
print("a = %d" %a)
print("b = %d" %b)
print("c = %d" %c)

```

```python
# Homework 2.2
# initial function
def fct(x) :
    return np.sin(x*math.pi)

# Exact Function - Differential Equation from the original
def fct_first_diff(x) :
    return math.pi * np.cos(x*math.pi)


# a forward - FDM
def forward_fdm (f,t,h) :
    return ((f(t + h) - f(t)) / h )


#b central - FDM
def central_fdm (f,t,h) :
    return ((f(t + h) - f(t - h)) / ( 2 * h) )



#c Higer order FDM
def high_order_fdm (f,t,h) :
    return  (  f(t - (2*h) )  - 8*f(t - h)  + 8*f(t + h)   - f(t + (2*h)) / ( 12 * h)            )



# using h = 0.1 / t = 0.4
# to call function and stores the calculated value to assigned variables

# assigning variable for Finite Diffent method
ans_forward_fdm = forward_fdm(fct,0.4,0.1)
ans_central_fdm = central_fdm(fct,0.4,0.1)
ans_higher_fdm = high_order_fdm(fct,0.4,0.1)


# assigning variable for first derivative function
ans_first_diff = fct_first_diff(0.4)


# Compare result 
#answer a
error_forward_fdm      = np.abs(ans_forward_fdm - ans_first_diff)

# answer b central - FDM
error_central_fdm       = np.abs(ans_central_fdm - ans_first_diff)


# answer c  Higer order FDM
error_higher_order_fdm = np.abs(ans_higher_fdm - ans_first_diff)

```

```python
# Answer Homework 2.2
print("--------------------------------------------------------------------")
print("from function u = sin(pi * X) / given x = 0.4 and delta x = 0.1\n")
print("----------------------this is calculated value----------------------\n")
print("Solution from Exact Equation  = %.4f" %ans_first_diff)
print("--------------------------------------------------------------------")
print("Solution from Forward FDM         = %.4f"  %ans_forward_fdm)
print("Solution from Central FDM         = %.4f"  %ans_central_fdm)
print("Solution from Higer Order FDM     = %.4f"  %ans_higher_fdm)
print("--------------------------------------------------------------------")
print("Error between Forward/Exact       = %.4f"%error_forward_fdm)
print("Error between Central/Exact       = %.4f"%error_central_fdm)
print("Error between Higher Order/Exact  = %.4f"%error_higher_order_fdm)
print("--------------------------------------------------------------------")

```
