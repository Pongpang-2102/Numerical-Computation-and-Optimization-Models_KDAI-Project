# In-Class Exercise

## 1. Forward Finite Difference

### 1.1 Using sin(x)

```python
def fct(x) :
    return np.sin(x) # we'll use sin(x) to test . Find d sin(x) /dx

def forward (f,t,h) :
    return ((f(t + h) - f(t)) / h )


ans = forward(fct,0,0.000001)
print(ans)  # our answer from forward would be 0.9999999999998334
```

### 1.2 Usingg tan(x)

```python
def fct(x) :
    return np.tan(x) # we'll use sin(x) to test . Find d sin(x) /dx

def forward (f,t,h) :
    return ((f(t + h) - f(t)) / h )


ans_forward = forward(fct,0,0.000001)
#ans = forward(fct,0,0.0001)
print(ans_forward)
```

## 2. Backward Finite Difference

### Using tan(x)

```python
def backward (f,t,h) :
    return ((f(t) - f(t - h)) / h )

ans_backward_bigO_4 = backward(fct,0,0.0001)
ans_backward_bigO_6 = backward(fct,0,0.000001)

print(ans_backward_bigO_4)
print(ans_backward_bigO_6 )
```

```python
# initial function
def fct(x) :
    return np.sin(x*math.pi)

# Exact Function - Differential Equation from the original
def fct_first_diff(x) :
    return math.pi * np.cos(x*math.pi)


# a forward - FDM
def forward_fdm (f,t,h) :
    return ((f(t + h) - f(t)) / h )


#b central - FDM
def central_fdm (f,t,h) :
    return ((f(t + h) - f(t - h)) / ( 2 * h) )


# using h = 0.1 / t = 0.4
# to call function and stores the calculated value to assigned variables

# assigning variable for Finite Diffent method
ans_forward_fdm = forward_fdm(fct,0.4,0.1)
ans_central_fdm = central_fdm(fct,0.4,0.1)


# assigning variable for first derivative function
ans_first_diff = fct_first_diff(0.4)


# Compare result 
error_forward_fdm = np.abs(ans_forward_fdm - ans_first_diff)
error_central_fdm = np.abs(ans_central_fdm - ans_first_diff)

```
