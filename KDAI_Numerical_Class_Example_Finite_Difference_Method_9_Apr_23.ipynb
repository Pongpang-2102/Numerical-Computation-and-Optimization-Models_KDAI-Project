```python
# import package
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fct(x) :
    return np.sin(x)


def forward (f,t,h) :
    return ((f(t + h) - f(t)) / h )


ans1 = np.zeros(6)
i = 0 
h2 = [10,1,0.1,0.01,0.001,0.0001]

for h1 in h2 :
    ans1[i] = forward(fct, 0 , h1)
    i = i + 1
print(ans1)


error = np.abs(ans1 - 1)
plt.plot(h2 ,ans1)

plt.plot(h2 ,error)

# naming the x axis
plt.xlabel('h')

#naming the y axis
plt.ylabel('y')

# add a legend
plt.legend(['numerical solution','error'])

# giving a title to my graph
plt.title('numerical solution and error')

plt.show()


```
